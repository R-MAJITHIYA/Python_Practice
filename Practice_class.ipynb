```python
class Abc():
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
        print("This is init of Abc.")

    def __str__(self):
        return "A = %d, B = %d, C = %d " % (self.a,self.b,self.c)

    def test(self):
        print("This is test of Abc.")

class Xyz(Abc):
    def __init__(self,x,y,z):
        Abc.__init__(self,x,y,z)    # init overriding.
        self.x = x
        self.y = y
        self.z = z
        print("This is init of Xyz.")

    def __str__(self):
        Abc.__str__(self)
        return "X = %d, Y = %d, Z = %d " % (self.x,self.y,self.z)

    def test(self):
        Abc.test(self)          #   super().test()
        print("This is test of Xyz.")

obj1 = Xyz(1,2,3)
print(obj1)
obj1.test()
```

```python
class A():       # Multi-level
    def __init__(self,a):
        self.a = a
    def __str__(self):
        return " A value is : %d. " % (self.a)
    def test(self):
        print('This is test of A.')

class B(A):
    def __init__(self,a,b):
        A.__init__(self,a)             #  if called by class name, it contains self
        self.b = b
    def __str__(self):
        #A.__str__(self)      str method not called as init method.
        return f"{A.__str__(self)}" + "\n B value is : %d. " % (self.b)
    def test(self):
        A.test(self)
        print('This is test of B.')

class C(B):
    def __init__(self,a,b,c):
        super().__init__(a,b)         # if called by super, it not contain self
        self.c = c
    def __str__(self):
        #super().__str__()      str method not called as init method.
        return f"{super().__str__()}" + "\n C value is : %d. " % (self.c)
    def test(self):
        super().test()
        print('This is test of C')

p1 = C(1,2,3)
print(p1)
p1.test()
```

```python
class A():       # Multiple
    def __init__(self,a):
        self.a = a
    def __str__(self):
        return " A value is : %d. " % (self.a)
    def test(self):
        print('This is test of A.')

class B():
    def __init__(self,b):
        self.b = b
    def __str__(self):
        return " B value is : %d. " % (self.b)
    def test(self):
        print('This is test of B.')

class C(A,B):
    def __init__(self,a,b,c):
        A.__init__(self,a)    # init is a private method and called by class name only
        B.__init__(self,b)
        self.c = c
    def __str__(self):
        return " C value is : %d. " % (self.c)
    def test(self):
        A.test(self)       # method can be called by object and class name
        B.test(self)
        print('This is test of C')

p1 = C(1,2,3)
print(p1)
p1.test()
'''if test method does not exist in child class C.
then it is called class A test method. Because A appear first in defination.
class C(A,B)  ->  A test
class C(B,A)  ->  B test'''
```
