```python
class Person:
    def __init__(self,id,name,dob):
      self.id = id
      self.name = name
      self.dob = dob
    def __str__(self):
      return "ID is : %d, Name is : %s and DOB is : %d" % (self.id, self.name, self.dob)
    def age(self, year):
      return year - self.dob

a = Person(1,'abc',1985)
print(a.id)
print(a.name)
print(a.dob)
print(a)
print(a.age(2022))
```

```python
class Triangle:
  def __init__(self,s,a,b,c):
    self.s = s
    self.a = a
    self.b = b
    self.c = c
  def __str__(self):
    return "Area of a Triangle is : %f." % ((self.s*(self.s-self.a)*(self.s-self.b)*(self.s-self.c))**5)

a = Triangle(5,2,1,4)
print(a)
```
